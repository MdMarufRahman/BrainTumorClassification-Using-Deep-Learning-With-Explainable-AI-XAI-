# Setup Image Paths and Labels

## For Train Set

```python
import numpy as np
import pandas as pd
import os

Id=[]
# Append directory name and filenames
for dirname, _, filenames in os.walk('/content/drive/MyDrive/BrainTumorClassification/Dataset/Train'):
    for filename in filenames:
        Id.append(os.path.join(dirname, filename))
Id[:10]
```

```python
# Convert filenames into dataframe
train = pd.DataFrame()
train = train.assign(filename=Id)
train.head()
```

```python
# Add new column 'label' and process labels for images
train['label'] = train['filename']

for i in range(len(train)):
    if 'glioma' in train['filename'][i]:
        train['label'][i] = 'glioma'
    if 'meningioma' in train['filename'][i]:
        train['label'][i] = 'meningioma'
    if 'notumor' in train['filename'][i]:
        train['label'][i] = 'notumor'
    if 'pituitary' in train['filename'][i]:
        train['label'][i] = 'pituitary'
```

```python
train['label'].value_counts()
```

## For Test Set

```python
Id = []
for dirname, _, filenames in os.walk('/content/drive/MyDrive/BrainTumorClassification/Dataset/Test'):
    for filename in filenames:
        Id.append(os.path.join(dirname, filename))
Id[:10]
```

```python
# Convert image id's to pandas dataframe
test = pd.DataFrame().assign(filename=Id)
test.head()
```

```python
test.tail()
```

```python
# Add new column 'label' and process labels for images
test['label'] = test['filename']

for i in range(len(test)):
    if 'glioma' in test['filename'][i]:
        test['label'][i] = 'glioma'
    if 'meningioma' in test['filename'][i]:
        test['label'][i] = 'meningioma'
    if 'notumor' in test['filename'][i]:
        test['label'][i] = 'notumor'
    if 'pituitary' in test['filename'][i]:
        test['label'][i] = 'pituitary'
```
